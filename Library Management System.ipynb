```python
Book_Record=[]
```

```python
def Addnewbook():
    Id=int(input("Enter Book Id:"))
    k=0
    for i in Book_Record:
        if(i[0]==Id):
            k=1
            print(k)
    if(k!=1):
        title=(input("Enter Book title:"))
        author=(input("Enter Book author:"))
        genre=(input("Enter Book genre:"))
        bookdetails=[Id,title,author,genre]
        Book_Record.append(bookdetails)
        print(Book_Record)
    else:
        print("Book Id Exists")
        
```

```python
def Removebook():
    Id=int(input("Enter Book Id:"))
    flag=True
    for i in Book_Record:
        if(i[0]==Id):
            Book_Record.remove(i)
            print("Record deleted ")
            flag=False
            break
    if(flag==True):
        print("Record Not Found")
    
```

```python
def Update_Book():
    Id=int(input("Enter Book Id:"))
    flag=True
    for i in Book_Record:
        if(i[0]==Id):
            print("Select the option to update data:\n 1.title\n2.author\n3.genre")
            ch=int(input())
            if(ch==1):
                title=input("Enter the title:")
                i[1]=title
            elif(ch==2):
                author=input("Enter the author:")
                i[2]=author
            elif(ch==3):
                genre=(input("Enter the genre:"))
                i[3]=genre
            print("Data Updated")
            flag=False
            break
    if(flag==True):
        print("Record Not Found")
    
```

```python
borrower=[]
def Borrow_book():
    Id=int(input("Enter Book Id:"))
    flag=True
    for i in Book_Record:
        if(i[0]==Id):
            print("Book is available:")
            if(len(i)==4):
                idborr=int(input("Enter the id of borrower"))
                nameborr=(input("Enter the name of borrower"))
                i.append([idborr,nameborr])
            else:
                print("Book Borrowed")
            flag=False
            break
    if(flag==True):
        print("Record Not Found")
```

```python
def Return_book():
    Id=int(input("Enter Book Id:"))
    flag=True
    for i in Book_Record:
        if(i[0]==Id):
            print("len:",(len(i)))
            print("Book is available:")
            if(len(i)==4):
                print("This Book is Not Borrowed")
                
            else:
                print(len(i))
                i.remove(i[4])
                print("Book returned")
            flag=False
            break
    if(flag==True):
        print("Record Not Found")
    
```

```python
def Menu():
    while True:
        print("----------------------------------------------------------")
        print(" Options available are:")
        print("1. Addnewbook")
        print("2. Removebook")
        print("3. Update_Book")
        print("4. Borrow_book")
        print("5. Return_book ")
        print("6. searchbook")
        print("7. AllBookdetails")
        print("8. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            Addnewbook()
        elif choice == "2":
            Removebook()
        elif choice == "3":
            Update_Book()
        elif choice == "4":
            Borrow_book()
        elif choice == "5":
            Return_book()
        elif choice == "6":
            searchbook()
        elif choice == "7":
            AllBookdetails()
        elif choice == "8":
            print("Exit")
            break
        else:
            print("Invalid choice. Please try again.")
```
