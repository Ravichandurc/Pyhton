```python
productdict = []

def addproduct():
    product_id = input("Product ID: ")
    name = input("Product name: ")
    price = float(input("Product price: "))
    description = input("Product description: ")
    stock = int(input("Product stock quantity: "))

    product_details = {
        "productid": product_id,
        "name": name,
        "price": price,
        "description": description,
        "stock": stock
    }

    productdict.append(product_details)
    print("Product added.")
def getproduct():
    product_id = input("Enter the product ID to retrieve: ")
    for product in productdict:
        if product["productid"] == product_id:
            return product
    return None

def getallproducts():
    return productdict


def updateproduct():
    product_id = input("Enter the product id to update: ")
    for product in productdict:
        if product["productid"] == product_id:
            product["name"] = input("Updated product name: ")
            product["price"] = float(input("Updated product price: "))
            product["description"] = input("Updated product description: ")
            product["stock"] = input("Updated product stock: ")
            print("Product updated successfully.")
            return
    print("Data not found.")

```

```python
def updatestock():
    product_id = input("Enter the product ID to update stock: ")
    quantity = int(input("Enter the quantity"))
    for product in productdict:
        if product["productid"] == product_id:
            product["stock"] += quantity
            print("Stock updated successfully.")
            return
    print("Product not found.")

def getstock():
    product_id = input("Enter the product ID to retrieve stock: ")
    for product in productdict:
        if product["productid"] == product_id:
            return product["stock"]
    return None
```

```python
orders = []

def placeorder():
    customer_id = input("Enter customer ID: ")
    products = []
    while True:
        product_id = input("Enter product ID (enter 'done' to finish): ")
        if product_id.lower() == "done":
            break
        quantity = int(input("Enter product quantity: "))
        product = {
            "productid": product_id,
            "quantity": quantity
        }
        products.append(product)
    order = {
        "order_id": len(orders) + 1,
        "customer_id": customer_id,
        "products": products,
        "status": "New"
    }
    orders.append(order)
    print("Order placed successfully.")

def updateorderstatus():
    order_id = int(input("Enter the order ID to update status: "))
    status = input("Enter the updated status: ")
    for order in orders:
        if order["order_id"] == order_id:
            order["status"] = status
            print("Order status updated successfully.")
            return
    print("Order not found.")

def getorder():
    order_id = int(input("Enter the order ID to retrieve: "))
    for order in orders:
        if order["order_id"] == order_id:
            return order
    return None

def getallorders():
    return orders
```
