```python
def Watchlist_Management():
    flag=True
    print(" Options available For search:")
    print("1. Add watch list")
    print("2. Remove From watch list")

    choice = input("Enter your choice: ")
    if choice =='1':
        Title=input("Enter the movie name:").lower()
        for i in movie_deatils1:
            if i["Title"]==Title:
                    Watch_list.append(i)
                    print("Movie Added ")
                    flag=False
    elif choice == "2":
        Title=input("Enter the movie name:").lower()
        if len(Watch_list)!=0:
            for i in Watch_list:
                if i["Title"]==Title:
                    Watch_list.remove(i)
                    print("Movie Removed ")
                    flag=False
        else:
            print("No movie in Watch list")
    if flag==True:
        print("No Data Found")
    
```
