```python
import numpy as np
from datetime import datetime,timedelta
Restaurants=[{'Restaurant':'Goandhara','Location':'koramangala','Menu':{'Biryani':250,'Dosa':50,'Idly':20,'pongal':30,'Rice':100}},
             {'Restaurant':'Meghanas','Location':'koramangala','Menu':{'Omlette':20,'rice':90,'dosa':40,'momo':350,'salad':120}},
             {'Restaurant':'Maven','Location':'Ejipura','Menu':{'Noodles':120,'vada':25,'salad':150,'Biryani':350,'Rice':75}},
              {'Restaurant':'Aromas','Location':'HSRlayout','Menu':{'momo':250,'upma':30,'pongal':35,'Omlette':35,'Rice':75}}
            ]
Orders=[]
Deliverypersons=['Revu','Vasanth','Abhi','lokha','shami','surya','Teja']
Avilable=['Revu','Vasanth','Abhi','lokha','shami','surya','Teja']
Busy=[]
DeliveryTime=[]
```

```python
def Restaurant_Search():
    print("Enter the criteria of searching:\n1.Location\n2.cuisine")
    flag=True
    choice=int(input())
    input1=input("Enter the Input:")
    for i in Restaurants:
        if(choice==1):
            if(i['Location']==(input1)):
                print("Restaurant:",i['Restaurant'])
                print("Location:",i['Location'])
                print("Menu:")
                for k,l in i['Menu'].items():
                    print(k,":",l)
                    
                #print(i['Menu'],end='')
                flag=False
        elif(choice==2):
            for j in i['Menu'] :
                #print(j)
                if(j==input1):
                    print("Restaurant:",i['Restaurant'])
                    print("Location:",i['Location'])
                    print("Menu:")
                    for k,l in i['Menu'].items():
                        print(k,":",l)
                flag=False
        else:
            print("Select from above options")
    if(flag==True):
        print("Record Data not Found")
```

```python
def placing_order():
    flag1=True
    flag2=True
    cost=0
    Restaurant=input("Enter the name of Restaurant :")
    Item=input("Enter the item from menu :")
    
    for i in Restaurants:
        if(i['Restaurant']==Restaurant):
            flag1=False
            res=i['Menu']
            for j in i['Menu']:
                if j==Item:
                    cost=res[Item]
                    #print(cost)
                    flag2=False
                    print("Item available")
    if(flag1==True):
        print("Restaurant not Found")
    if(flag2==True):
        print("Dish Not Found")
    if flag1==False and flag2==False:
        order_id=np.random.randint(100000000)
        Quantity=int(input("Enter Quantity:"))
        price=cost*Quantity
        Name=input("Enter Your name:")
        phonenumber=input("Enter Phone Number:")
        addres=input("Enter addres:")
        Ordered_time = datetime.now()
        O_time=Ordered_time.strftime("%H:%M:%S")
        result = Ordered_time + timedelta(minutes=50)
        Estimate_Delivery_time = result.strftime("%H:%M:%S")
        Status='Ordered'
        if(len(Avilable)!=0):
            Deliveryperson=Avilable[0]
            print(Deliveryperson)
            Busy.append(Avilable[0])
            Avilable.remove(Avilable[0])
            print(Avilable)
            print(Busy)
            
        else:
            print("No delivery person available:")
            return 0
        Orders.append([order_id,Restaurant,Item,Quantity,price,Deliveryperson,O_time,Estimate_Delivery_time,Status,Name,phonenumber,addres])
    #print(Orders)
    print("Order_id:",order_id)
    print("Restaurant_Name:",Restaurant)
    print("Dish ordered:",Item)
    print("Quantity:",Quantity)
    print("Cost:",price)
    print("Deliveryperson name:",Deliveryperson)
    print("Ordered Time:",O_time)
    print("Estimate_Delivery_time:",Estimate_Delivery_time)
    print('Status:',Status)
    print('---your details----')
    print('Person Name:',Name)
    print("phonenumber of person:",phonenumber)
    print('addres of person:',addres)
            
```

```python
def Order_Track():
    Order_id=int(input("Enter order id:"))
    flag=True
    for i in Orders:
        if i[0]==Order_id:
            flag=False
            print("Order_id:",i[0])
            print("Restaurant_Name:",i[1])
            print("Dish ordered:",i[2])
            print("Quantity:",i[3])
            print("Cost:",i[4])
            print("Deliveryperson name:",i[5])
            print("Ordered Time:",i[6])
            print("Estimate_Delivery_time:",i[7])
            print('Status:',i[8])
            print('---your details----')
            print('Person Name:',i[9])
            print("phonenumber of person:",i[10])
            print('addres of person:',i[11])
            print("----------------------------------------------------------")
            
    if flag==True:
        print('Order Not Found')
```

```python
def Order_updation():
    Order_id=int(input("Enter order id:"))
    flag=True
    for i in Orders:
        if i[0]==Order_id:
            flag=False
            
            print("Enter order Status:")
            print("1. Delivered")
            print("2. Not Delivered")
            k=input()
            if k =='1':
                Order_staus='Delivered'
            elif k == "2":
                Order_staus='Not Delivered'
                return 0
            print(" Are you Willing to Provide Feed back :")
            print("1. yes")
            print("2. no")
            choice=input("Enter your Choice:")
            if choice =='1':
                Feedback=input('Enter FeedBack')
            elif choice == "2":
                Feedback='No feedback'  
            i[8]=Order_staus
            i.append(Feedback)
            if Order_staus=='Delivered':
                d_person=i[5]
                Busy.remove(d_person)
                Avilable.append(d_person)
                
            oo_time=i[6]
            present = datetime.now()
            result = present + timedelta(minutes=50)
            date = present.strftime("%H:%M:%S")
            result1 = result.strftime("%H:%M:%S")
    
            time_difference = (datetime.strptime(date, "%H:%M:%S") - datetime.strptime(oo_time, "%H:%M:%S")) % timedelta(days=1)
            difference_seconds = time_difference.total_seconds()

            print("Time difference:", time_difference)
            print("Time difference in seconds:", difference_seconds)
            DeliveryTime.append(time_difference)
    if flag==True:
        print('Order Not Found')
        
    
```

```python
def occurance_rate():
    dict1 = {}
    rc={}
    for i in Orders:
        value = i[1]
        if value in dict1:
            dict1[value] += 1
        else:
            dict1[value] = 1
        rcv=i[2]
        if rcv in rc:
            rc[rcv] += 1
        else:
            rc[rcv] = 1
    print("1",dict1)
    print("2",rc)    
    print('Returant Wise Report:')
    for key, value in dict1.items():
        print("Total orders for", key + ":", value)
    print('Dish Wise Report:')
    for key1, values in rc.items():
        print("Total orders for", key1 + ":", values)
    #print("Delivery times:",DeliveryTime)
    #timedeltas = [timedelta(seconds=43), timedelta(seconds=58)]
    if len(DeliveryTime)==0:
        print("No Data")
    else:
        total_timedelta =sum(DeliveryTime, timedelta())  # Initialize the total as timedelta 0
        average_timedelta = total_timedelta / len(DeliveryTime)
        print("Average time:", average_timedelta)
    
```

```python
def order_history():
    for i in Orders:
        print("Order_id:",i[0])
        print("Restaurant_Name:",i[1])
        print("Dish ordered:",i[2])
        print("Quantity:",i[3])
        print("Cost:",i[4])
        print("Deliveryperson name:",i[5])
        print("Ordered Time:",i[6])
        print("Estimate_Delivery_time:",i[7])
        print('Status:',i[8])
        print('---your details----')
        print('Person Name:',i[9])
        print("phonenumber of person:",i[10])
        print('addres of person:',i[11])
        print("----------------------------------------------------------")
```

```python
def re_order_from_history():
    flag=True
    Order_id=int(input("Enter order id:"))
    for i in Orders:
        if Order_id==int(i[0]):
            flag==False
            order_id=np.random.randint(100000000)
            price=i[4]*int(i[3])
            Ordered_time = datetime.now()
            O_time=Ordered_time.strftime("%H:%M:%S")
            result = Ordered_time + timedelta(minutes=50)
            Estimate_Delivery_time = result.strftime("%H:%M:%S")
            Status='Ordered'
            if(len(Avilable)!=0):
                Deliveryperson=Avilable[0]
                Busy.append(Avilable[0])
                Avilable.remove(Avilable[0])
                
            else:
                print("No delivery person available:")
                return 0
            
        Orders.append([order_id,i[1],i[2],i[3],i[4],Deliveryperson,O_time,Estimate_Delivery_time,Status,i[9],i[10],i[11]])
        print("Order placed ------")
    if flag== True:
        print("No order with given order id")
```

```python
def Delivery_person_details():
    print("Avilable Delivery Persons are:")
    for i in Avilable:
        print(i)
    print("Busy Delivery Persons are:")
    for i in Busy:
        print(i)
    
        
```

```python
def Menu():
    while True:
        print("----------------------------------------------------------")
        print(" Options available are:")
        print("1. Restaurant_Search")
        print("2. placing_order")
        print("3. Order_Track")
        print("4. order_history ")
        print("5. re_order_from_history")
        print("6. Order_updation ")
        print("7. occurance_rate")
        print("8. Delivery_person_details")
        print("9. Exit")
        print("----------------------------------------------------------")
        choice = input("Enter your choice: ")
        if choice =='1':
            Restaurant_Search()
        elif choice == "2":
            placing_order()
        elif choice == "3":
            Order_Track()
        elif choice == "4":
            order_history()
        elif choice == "5":
            re_order_from_history()
        elif choice == "6":
            Order_updation()
        elif choice == "7":
            occurance_rate()
        elif choice == "8":
            Delivery_person_details()
        elif choice == "9":
            print("Exit")
            break
        else:
            print("Invalid choice. Please try again.")
```
